# Exploratory Data Analysis

#### Question 1. :- Find Number of States in This dataset?

```python
No_Of_States = d['State'].nunique()
print('There are %d States in this DataSet'%No_Of_States)
```

#### Question 2. :- Find Number of Cities in This dataset?

```python
No_Of_Cities = d['City'].nunique()
print('There are %d Cities in this DataSet'%No_Of_Cities)
```

#### Question 3. :- Find Number of Sub Category in This dataset?

```python
No_Of_SubCategory = d['Sub-Category'].nunique()
print('There are %d Sub-Category in this DataSet'%No_Of_SubCategory)
```

#### Question 4. :- Which ship mode is Highly preferred?

```python
# .reset_index() = used to convert series to datafram
# ship_mode_count = d['Ship Mode'].value_counts()
ship_mode_count = d['Ship Mode'].value_counts().reset_index() 
ship_mode_count
```

```python
sns.countplot(x=d['Ship Mode'])
```

- We can say that Standard Class ship mode is highly preferred.

#### Question 5. :- Which Segment type is Highly preferred?

```python
Segment_count = d['Segment'].value_counts().reset_index() 
Segment_count
```

```python
sns.countplot(x=d['Segment'])
```

- We can say that Consumer type Segment is highly preferred.

#### Question 5. :-Find top 5 Cities which has highest market?

```python
City_count = d['City'].value_counts().reset_index().head(5)
City_count
```

- We can say that New York City has highest market

#### Question 6. :- In which Region has highest market?

```python
Region_count = d['Region'].value_counts().reset_index()
Region_count
```

```python
sns.countplot(x=d['Region'])
```

- We can say that West Region has highest market

#### Question 7. :- Which Category of product has highest market?

```python
Category_count = d['Category'].value_counts().reset_index()
Category_count
```

```python
sns.countplot(x=d['Category'])
```

#### Question 8. :- Which Category of product is Highly sold?

```python
# To fetch 2 Columns d.groupby('Category')['Sales']
Category_sold = d.groupby('Category')['Sales'].mean().reset_index().sort_values('Sales',ascending=False)
Category_sold
```

```python
sns.barplot(x=Category_sold['Category'],y=Category_sold['Sales'])
```

- We can say that Technology Category has highest sales

#### Question 9. :- Which Category of product has highest profit?

```python
Category_profit = d.groupby('Category')['Profit'].sum().reset_index().sort_values('Profit',ascending=False)
Category_profit
```

```python
sns.barplot(x=Category_profit['Category'],y=Category_profit['Profit'])
```

- We can say that Technology Category has highest profit

#### Question 10. :- Which Sub-Category of product is Highly sold?

```python
Sub_Category_sold = d.groupby('Sub-Category')['Sales'].sum().reset_index().sort_values('Sales',ascending=False)
Sub_Category_sold
```

```python
sns.barplot(x=Sub_Category_sold['Sub-Category'],y=Sub_Category_sold['Sales'])
```

## Comparison
#

## 1. State vs Sales

```python
sales_state = d.groupby(['State'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State' , y= 'Sales')
plt.title('State Vs Sales')
#plt.xlabel('State')
#plt.ylabel=('Sales')
```

- Above Graph is showing that which State has Highest Sales.
- Top 10 Highest Sales According to State wise.
- We can say that California State has Highest Sales and then New York and then Texas and so on..

## 2. State vs Profit

```python
sales_state = d.groupby(['State'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Profit')
```

- Above Graph is showing that which State has Highest Profit.
- Top 10 Highest Profit According to State wise.
- We can say that California State has Highest Profit and then New York and then Washington and so on

## 3. City vs Profit

```python
sales_city = d.groupby(['City'], as_index=False)['Profit'].sum().sort_values(by='Profit', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_city, x = 'City',y= 'Profit')
```

- Above Graph is showing that which City has Highest Profit.
- Top 10 Highest Profit According to City wise.
- We can say that New York City has Highest Profit and then Los Angeles and then Seattle and so on.

## 4. Region vs Sales

```python
sales_Region = d.groupby(['Region'], as_index=False)['Sales'].sum().sort_values(by='Sales', ascending=False).head()

sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_Region, x = 'Region',y= 'Sales')
```

- Above Graph is showing that which Region has Highest Sales.
- Top 4 Highest Sales According to Region wise.
- We can say that West Region has Highest Sales and then East and so on.
